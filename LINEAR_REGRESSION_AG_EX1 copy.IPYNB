```python
import numpy as np;
```

Glossary/Notes:
1. A seed makes randomness reproducible
2. y = (true linear relation) + (random noise) or Gaussian noise

```python
rng = np.random.default_rng(seed=42)
m = 200 # number of instances
x = 2 * rng.random((m,1)) # column vector (values between 0 and 2)
y = 4 + 3 * x + rng.standard_normal((m,1)) #column vector
```

```python
from sklearn.preprocessing import add_dummy_feature
```

```python
# Scatter plot of x vs y
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='blue', edgecolor='k')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter plot of x vs y')
plt.grid(True)
plt.show()
```

Glossary/Markdown:

 1. x_b = add_dummy_feature(x)

👉 This adds a column of 1s to your input data.

2. theta_best = Normal Equation (calculate parameters (bias & weight) without using gradient descent)


```python
x_b = add_dummy_feature(x)
theta_best = np.linalg.inv(x_b.T @ x_b) @ x_b.T @ y

theta_best
```

```python
x_new = np.array([[0],[2]])
x_new_b = add_dummy_feature(x_new)

y_predict = x_new_b @ theta_best
y_predict
```

```python
# Plot the original data points and the model predictions
plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='blue', label='Original data', alpha=0.7)
plt.plot(x_new, y_predict, color='red', linewidth=2, label='Prediction line')
plt.scatter(x_new, y_predict, color='red', marker='o', s=80, label='Predicted points')
plt.xlabel('x')
plt.ylabel('y')
plt.title('x vs y with predicted line')
plt.legend()
plt.grid(True)
plt.show()
```
